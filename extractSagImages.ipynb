```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import glob
from PIL import Image
from tqdm import tqdm
import timm
import pickle

DATA_PATH = "D:/ML/RSNA2024"
```

```python

dfDescr = pd.read_csv(os.path.join(DATA_PATH, "train_series_descriptions.csv"))
uniqueSeriesDescr = np.array(['Sagittal T2/STIR', 'Sagittal T1', 'Axial T2'])
# dfDescr.set_index(["study_id", "series_description"], inplace=True)
dfDescr.sort_index(inplace=True)
dfDescr.head(10)
```

```python
sagT1studyIds = dfDescr.loc[dfDescr["series_description"]=="Sagittal T1"]["study_id"].to_numpy()
sagT1seriesIds = dfDescr.loc[dfDescr["series_description"]=="Sagittal T1"]["series_id"].to_numpy()
sagT2studyIds = dfDescr.loc[dfDescr["series_description"]=="Sagittal T2/STIR"]["study_id"].to_numpy()
sagT2seriesIds = dfDescr.loc[dfDescr["series_description"]=="Sagittal T2/STIR"]["series_id"].to_numpy()

import random

sagT1data= list(zip(sagT1studyIds, sagT1seriesIds))
random.shuffle(sagT1data)

sagT2data= list(zip(sagT2studyIds, sagT2seriesIds))
random.shuffle(sagT2data)
```

```python
from utils import dicomToArray
from PIL import Image

studyCount=20

for (studyId, seriesId) in tqdm(sagT1data[0:studyCount], smoothing=0.05, total=studyCount):
    os.makedirs(os.path.join(DATA_PATH, f"sagImages/{studyId}"), exist_ok=True)

    allFilesInDir = glob.glob(os.path.join(DATA_PATH, f"train_images/{studyId}/{seriesId}/*.dcm"))
    #Sort by filename number
    allFilesInDir = sorted(allFilesInDir, key=lambda x:int(re.findall(r"\d+\.dcm", x)[0].replace(".dcm", "")))
    for f in allFilesInDir:
        arr = dicomToArray(f, (420,420))
        im = Image.fromarray(arr)
        head, filename = os.path.split(f)
        im.save(os.path.join(DATA_PATH, f"sagImages/{studyId}/T1_{filename.split(".dcm")[0]}.png"))

for (studyId, seriesId) in tqdm(sagT2data[0:studyCount], smoothing=0.05, total=studyCount):
    os.makedirs(os.path.join(DATA_PATH, f"sagImages/{studyId}"), exist_ok=True)


    allFilesInDir = glob.glob(os.path.join(DATA_PATH, f"train_images/{studyId}/{seriesId}/*.dcm"))
    #Sort by filename number
    allFilesInDir = sorted(allFilesInDir, key=lambda x:int(re.findall(r"\d+\.dcm", x)[0].replace(".dcm", "")))
    for f in allFilesInDir:
        arr = dicomToArray(f, (420,420))
        im = Image.fromarray(arr)
        head, filename = os.path.split(f)
        im.save(os.path.join(DATA_PATH, f"sagImages/{studyId}/T2_{filename.split(".dcm")[0]}.png"))
```
